## <span style="color:#4361EE">El Perceptrón</span>

- Imagina que un perceptrón es como un amigo robot en tu clase. Este amigo tiene una tarea muy especial: decidir si debe llevar paraguas o no al salir de la escuela. El robot toma esta decisión basándose en dos cosas que observa: si está lloviendo y si hay nubes oscuras en el cielo.

- Si cualquiera de estas dos cosas es cierta (está lloviendo o hay nubes oscuras), el robot decide llevar un paraguas. Si ninguna es cierta, no lleva paraguas. El robot, como un perceptrón, está tomando una decisión simple basada en la información que recibe.

## <span style="color:#4361EE">Redes Neuronales</span>

- Ahora, imagina que en lugar de un amigo robot, tienes un equipo de robots. Cada uno de ellos es como el amigo que decide sobre el paraguas, pero cada uno observa cosas diferentes. Uno puede mirar si está lloviendo, otro puede ver si hay nubes oscuras, otro puede sentir si hace frío, y otro puede incluso revisar el pronóstico del tiempo en internet.

- Todos estos robots luego le cuentan sus observaciones a un robot líder. El robot líder, después de escuchar a todos, toma la decisión final sobre llevar o no llevar paraguas. Esta es una red neuronal: un grupo de perceptrones (los amigos robots) que trabajan juntos para tomar una decisión más compleja, basada en más información.

## <span style="color:#4361EE">Funciones de Activación</span>

- Imagina que los robots amigos tienen un sistema de luces para decidir si llevar paraguas o no. Si la luz se enciende en verde, significa "llevar paraguas", y si se enciende en rojo, significa "no llevar paraguas".

- La función de activación es como una regla que el robot usa para decidir qué color de luz encender. Por ejemplo, podría ser algo como "si veo lluvia o nubes oscuras, enciendo la luz verde". Esta regla ayuda al robot a tomar una decisión clara basada en lo que observa, igual que la función de activación ayuda a un perceptrón a decidir qué hacer.

## <span style="color:#4361EE">Backpropagation</span>

Imagina que los robots amigos que deciden si llevar paraguas o no están organizados en un equipo. Al frente está el robot líder que toma la decisión final, y detrás de él hay otros robots que le dan información sobre el clima.

Un día, el robot líder toma una mala decisión: decide no llevar paraguas, pero resulta que llueve. El equipo de robots quiere aprender de este error para no repetirlo. Aquí es donde entra en juego la backpropagation, o "retropropagación" en español.

1. **Retroalimentación**:

- Primero, el robot líder dice: "He cometido un error, no debí decidir dejar el paraguas". Esta información es como una retroalimentación que el robot líder comparte con los otros robots.

2. **Ajustes en el Equipo**:

- Luego, cada robot en la fila piensa en cómo contribuyó a la decisión y qué podría hacer mejor. Por ejemplo, el robot que observa las nubes podría pensar: "Debería haber dado más importancia a esas nubes oscuras". Así, cada robot hace pequeños ajustes en su forma de observar y pensar sobre el clima.

3. **Mejorando Juntos**:

- Al hacer estos ajustes y compartir la retroalimentación a lo largo del equipo, todos los robots aprenden del error. La próxima vez, trabajarán mejor juntos para tomar una decisión más acertada sobre el paraguas.

## <span style="color:#4361EE">Clasificación de Imágenes</span>

Imagina que tienes una máquina mágica (una red neuronal) que quiere aprender a distinguir entre fotos de perros y gatos. Vamos a enseñarle cómo hacerlo paso a paso:

1. **Construyendo la Máquina Mágica:** Primero, creamos nuestra máquina mágica. Piensa en ella como un equipo de detectives (los nodos) que son muy buenos para aprender y recordar cosas, especialmente para saber si una imagen es de un perro o de un gato.

2. **Primer Caso - El Misterio del Perro:** Ahora, mostramos a nuestra máquina una imagen de un perro. Cada detective da su opinión: ¿Es un perro o un gato? Al principio, pueden estar confundidos y no acertar.

3. **Aprendiendo del Error:** Después de que los detectives adivinen, les decimos la respuesta correcta: "¡Es un perro!" Si se equivocaron, ahora sabrán un poco más sobre cómo se ve un perro.

4. **Practicando con el Perro:** Seguimos mostrando la misma imagen del perro muchas veces. Cada vez, los detectives intentan adivinar y aprenden de sus errores, mejorando su habilidad para reconocer perros.

5. **Segundo Caso - El Enigma del Gato:** Ahora hacemos lo mismo con una imagen de un gato. Los detectives adivinan, les decimos la respuesta correcta, y aprenden de sus errores.

6. **¡Detectives Expertos!** Después de mucha práctica con muchas imágenes de perros y gatos, nuestros detectives se vuelven expertos en distinguir entre ellos. Si les mostramos una nueva foto de un perro o un gato, ¡pueden decirnos con mucha precisión cuál es cuál!

**En Resumen:** Nuestra máquina mágica, como un grupo de detectives aprendices, se entrena viendo muchas fotos de perros y gatos. Con cada foto que ven, se vuelven mejores en saber qué animal es, ¡igual que tú te vuelves mejor en un juego cuanto más lo practicas!

## <span style="color:#4361EE">Convolución</span>

Una convolución es como una especie de "filtro mágico" que ayuda a una computadora a ver cosas en imágenes o detectar patrones.

Imagina que tienes un montón de stickers con formas especiales, como corazones, estrellas y círculos. Ahora, también tienes una imagen grande llena de muchos puntos de colores. Cuando pones un sticker sobre la imagen y lo deslizas a lo largo de ella, solo verás los puntos que coinciden con la forma del sticker. Es como si solo pudieras ver las partes de la imagen que se parecen a la forma del sticker.

La convolución hace algo similar en una computadora. Toma una pequeña ventana llamada "filtro" (como el sticker) y la desliza sobre una imagen. Cuando el filtro encuentra partes de la imagen que se parecen a lo que está buscando, resalta esas partes. Esto es útil para que la computadora detecte cosas como bordes de objetos o características importantes en una imagen.

#### Importando librerías

```python
import keras
from keras.layers import Dense, Dropout, Conv2D, MaxPool2D, Flatten, Activation
from keras.models import Model
from keras.optimizers import Adam
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from keras.callbacks import ModelCheckpoint, EarlyStopping

import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

#### Creacion de la Clase 

```python
class Cat_Dog(Model):
    def __init__(self, input_shape):
        super(Cat_Dog, self).__init__()
        self.conv1 = Conv2D(32, (3, 3), activation = 'relu', padding = 'same', input_shape = input_shape)
        self.conv2 = Conv2D(64, (3, 3), activation = 'relu', padding = 'same')
        self.conv3 = Conv2D(128, (3, 3), activation = 'relu', padding = 'same')
        self.maxpooling = MaxPool2D((2, 2))
        self.flatten = Flatten()
        self.drop = Dropout(0.4)
        self.d1 = Dense(512, activation = 'relu')
        self.d2 = Dense(1, activation = 'sigmoid')
        
    def call(self, x):
        x = self.conv1(x)
        x = self.maxpooling(x)
        x = self.conv2(x)
        x = self.maxpooling(x)
        x = self.conv3(x)
        x = self.maxpooling(x)
        x = self.flatten(x)
        x = self.drop(x)
        x = self.d1(x)
        x = self.d2(x)
        
        return x
```

#### Data augmentation

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
    )

test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
pepa = load_img('./Images/pepa.jpeg')
pepa_x = img_to_array(pepa)
pepa_x = pepa_x[400:1400, ]
pepa_x = pepa_x.reshape((1,)+ pepa_x.shape)
```

```python
batches = []
i = 0

for batch in train_datagen.flow(pepa_x, batch_size=1):
    batches.append(array_to_img(batch[0]))
    i += 1
    if i % 10 == 0:
        break


num_filas = int(np.ceil(len(batches) / 5))
num_columnas = 5

fig, axs = plt.subplots(num_filas, num_columnas, figsize=(12, 12))

for i in range(len(batches)):
    fila = i // num_columnas
    columna = i % num_columnas
    axs[fila, columna].imshow(batches[i])
    axs[fila, columna].axis('off')
    
plt.tight_layout()
plt.show()
```

#### Reescalar los datos

```python
train_generator = train_datagen.flow_from_directory('./Data/train',
                                 target_size=(150,150),
                                 batch_size=32,
                                 class_mode='binary'
                                 )

validation_generator = test_datagen.flow_from_directory('./Data/validation',
                                 target_size=(150,150),
                                 batch_size=32,
                                 class_mode='binary'
                                 )
```

```python
checkpoint = ModelCheckpoint('./Model', 
                             monitor = 'val_accuracy', 
                             verbose = 1, 
                             save_best_only = True, 
                             save_format='tf')
```

```python
model = Cat_Dog((150, 150, 3))
```

```python
model.compile(loss='binary_crossentropy', 
              optimizer = Adam(),
              metrics=['accuracy'])
```

```python
hist = model.fit(
    train_generator,
    steps_per_epoch = 2000 // 32,
    epochs = 100,
    validation_data = validation_generator,
    validation_steps = 1000 // 32,
    callbacks = [checkpoint]
)
```

```python
pd.DataFrame(hist.history)
```

```python
plt.plot(hist.history['loss'], label = 'Entrenamiento')
plt.plot(hist.history['val_loss'], label = 'Validacion')
plt.legend()
plt.show()
```

```python
plt.plot(hist.history['accuracy'], label = 'Entrenamiento')
plt.plot(hist.history['val_accuracy'], label = 'Validacion')
plt.legend()
plt.show()
```
